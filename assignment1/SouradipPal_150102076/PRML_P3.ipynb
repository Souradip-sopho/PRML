# Polynomial Regression on Income Survey Data(Atlantic Region,USA)

This is the code for problem P3.

In this problem,our aim is to find an association between the different attributes of the employees like age,calender year,education with the income earned by them.For this,polynomial regression is performed individually on each of these attributes.For finding the appropriate value of the degree of polynomial features to be used,regression was done using different degrees less than 100 and the optimal degree is taken where the training error(mean-square-error) is least.
We have used the matrix form of polynomial regression model defined as follows:<br>
$$r=A\omega +\epsilon$$ where $r$ is the target vector,$A$ is the design matrix,$\omega$ is the weight vector and $\epsilon$ is the error vector.<br>
The optimal weight vector is calculated  as follows:$$\omega^*= ((A^TA)^{-1})A^Tr$$

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

For easier processing of data, we have modified the file containing the income data by adding the correspoing feature titles to the file and renaming the file as 'Wage_dataset.csv'.Accordingly,the dataset was loaded into a data frame using pandas library. 

```python
data = pd.read_csv("Wage_dataset.csv")
year = np.array(data["year"])
age = np.array(data["age"])
education = np.array(data["education"])
wage = np.array(data["wage"])

size=wage.shape
```

## Case 1 : Feature - Age
The following code computes the polynomial regression on the wage data using the attribute age.Also shown the predicted labels on the training data over the scatter plot of the data.

```python
### Feature: Age ###
min_loss_age = math.inf
optimal_n = 0
print("Variation of Error with degree of polynomial(age)")
plt.figure(figsize=(10, 10))
for n in range(1,100):
    age_matrix = np.zeros((size[0],n))
    for i in range(size[0]):
        for j in range(n):
            age_matrix[i][j] = age[i]**j
    age_weight = np.matmul(np.linalg.pinv(age_matrix),wage)
    predicted_wage_age = np.matmul(age_matrix,age_weight)
    error=np.subtract(predicted_wage_age,wage)
    loss_age=np.dot(error.T,error)
    plt.plot(n,loss_age,'ro')
    if (loss_age < min_loss_age):
        min_loss_age = loss_age
        optimal_n = n

plt.show()
print("Optimal degree of polynomial(degree<100): ",end="")
print(optimal_n)
```

```python
print("Scatter Plot(Wage Vs Age): ")
plt.figure(figsize=(10, 10))
plt.scatter(age,wage)
plt.scatter(age,predicted_wage_age,c='g')
plt.show()
```

Here, the predicted wages are plotted for each training sample for the optimal degree of polynomial for the feature age.

```python
N = optimal_n
age_matrix = np.zeros((size[0],N))
loss_array_age = np.zeros((size[0],1))
for i in range(size[0]):
    for j in range(N):
        age_matrix[i][j] = age[i]**j
age_weight = np.matmul(np.linalg.pinv(age_matrix),wage)
predicted_wage_age = np.matmul(age_matrix,age_weight)
x = np.arange(size[0])
print("Plot showing Predicted Wages(Actual Wages->Blue,Predicted Wages->Red):")
plt.figure(figsize=(30, 10))
plt.plot(x,wage,marker='o')
plt.plot(x,predicted_wage_age,marker='x',c='r')
plt.show()
error_age=np.subtract(predicted_wage_age,wage)
loss_age=np.dot(error_age.T,error_age)
print("Minimun loss using feature age for optimal degree of polynomial (degree<100): ",end="")
print(loss_age)
```

## Case 2 : Feature - Calender Year
The following code computes the polynomial regression on the wage data using the attribute year.

```python
### Feature: Year ###
min_loss_year = math.inf
optimal_m = 0
print("Variation of Error with degree of polynomial(year)")
plt.figure(figsize=(10, 10))
for m in range(1,100):
    year_matrix = np.zeros((size[0],m))
    for i in range(size[0]):
        for j in range(m):
            year_matrix[i][j] = year[i]**j
    year_weight = np.matmul(np.linalg.pinv(year_matrix),wage)
    predicted_wage_year = np.matmul(year_matrix,year_weight)
    error_year=np.subtract(predicted_wage_year,wage)
    loss_year=np.dot(error_year.T,error_year)
    plt.plot(m,loss_year,'ro')
    if (loss_year < min_loss_year):
        min_loss_year = loss_year
        optimal_m = m

plt.show()
print("Optimal degree of polynomial(degree<100): ",end="")
print(optimal_m)
```

```python
print("Scatter Plot(Wage Vs Year): ")
plt.figure(figsize=(10, 10))
plt.scatter(year,wage)
plt.scatter(year,predicted_wage_year,c='g')
plt.show()
```

```python
M=optimal_m
year_matrix = np.zeros((size[0],M))
loss_array_year = np.zeros(size[0])
for i in range(size[0]):
    for j in range(0,M):
        year_matrix[i][j] = year[i]**j

year_weight = np.matmul(np.linalg.pinv(year_matrix),wage)
predicted_wage_year = np.matmul(year_matrix,year_weight)
x = np.arange(size[0])
print("Plot showing Predicted Wages(Actual Wages->Blue,Predicted Wages->Red):")
plt.figure(figsize=(30, 10))
plt.plot(x,wage,marker='o')
plt.plot(x,predicted_wage_year,marker='x',c='r')
plt.show()
error_year=np.subtract(predicted_wage_year,wage)
loss_year=np.dot(error_year.T,error_year)
print("Minimun loss using feature year for optimul degree of polynomial (degree<100): ",end="")
print(loss_year)
```

## Case 3 : Feature - Education
The following code computes the polynomial regression on the wage data using the attribute education.

```python
### Feature: Education ###
min_loss_education = math.inf
optimal_l = 0
print("Variation of Error with degree of polynomial(education)")
plt.figure(figsize=(10, 10))
for l in range(1,100):
    education_matrix = np.zeros((size[0],l))
    for i in range(size[0]):
        for j in range(0,l):
            education_matrix[i][j] = education[i]**j
    education_weight = np.matmul(np.linalg.pinv(education_matrix),wage)
    predicted_wage_education = np.matmul(education_matrix,education_weight)
    error_education=np.subtract(predicted_wage_education,wage)
    loss_education=np.dot(error_education.T,error_education)
    plt.plot(l,loss_education,'ro')
    if (loss_education < min_loss_education):
        min_loss_education = loss_education
        optimal_l = l
plt.show()
print("Optimal degree of polynomial(degree<100): ",end="")
print(optimal_l)
```

```python
print("Scatter Plot(Wage Vs Education): ")
plt.figure(figsize=(10, 10))
plt.scatter(education,wage)
plt.scatter(education,predicted_wage_education,c='g')
plt.show()
```

```python
L=optimal_l
education_matrix = np.zeros((size[0],L))
loss_array_education = np.zeros(size[0])
for i in range(size[0]):
    for j in range(0,L):
        education_matrix[i][j] = education[i]**j
education_weight = np.matmul(np.linalg.pinv(education_matrix),wage)
predicted_wage_education = np.matmul(education_matrix,education_weight)
x = np.arange(size[0])
print("Plot showing Predicted Wages(Actual Wages->Blue,Predicted Wages->Red):")
plt.figure(figsize=(30, 10))
plt.plot(x,wage,marker='o')
plt.plot(x,predicted_wage_education,marker='x',c='r')
plt.show()
error_education=np.subtract(predicted_wage_education,wage)
loss_education=np.dot(error_education.T,error_education)
print("Minimun loss using feature education for optimul degree of polynomial (degree<100): ",end="")
print(loss_education)
```

It is seen from the above experiment that a single attribute cannot be used to predict a employee's wage as multiple examples contains the same age or year or education attribute but different wages.As some of the attributes take only discrete values,different target values are obtained for the same value of the attribute.It  is therefore not a good choice to use polynomial regression using a single feature.Thus multiple features must be used for polynomial regression for this data.
